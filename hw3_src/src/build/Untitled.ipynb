```python
import matplotlib.pyplot as plt

# Lists to store the data
arr_rates = list(range(10, 20))
exp_avg_response_time = []
dist1_avg_response_time = []

# Read data from saved text files and extract response times
for arr_rate in arr_rates:
    with open(f'exp_arr_{arr_rate}.txt', 'r') as exp_file:
        lines = exp_file.readlines()
        response_times = []
        for line in lines:
            if line.startswith("R"):
                parts = line.split(",")
                if len(parts) >= 2:
                    response_time = float(parts[0].split(":")[-1])
                    response_times.append(response_time)
        exp_avg_response_time.append(sum(response_times) / len(response_times))  # Calculate average response time

    with open(f'dist1_arr_{arr_rate}.txt', 'r') as dist1_file:
        lines = dist1_file.readlines()
        response_times = []
        for line in lines:
            if line.startswith("R"):
                parts = line.split(",")
                if len(parts) >= 2:
                    response_time = float(parts[0].split(":")[-1])
                    response_times.append(response_time)
        dist1_avg_response_time.append(sum(response_times) / len(response_times))  # Calculate average response time

# Create the plot
plt.plot(arr_rates, exp_avg_response_time, label='Exponential Distribution (-d 0)')
plt.plot(arr_rates, dist1_avg_response_time, label='Distribution 1 (-d 1)')
plt.xlabel('Arrival Rate')
plt.ylabel('Average Response Time')
plt.legend()
plt.title('Quality of Service Comparison')
plt.grid(True)

# Save the plot as an image file
plt.savefig('qos_comparison.png')

# Show the plot
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

# Load data
df = pd.read_csv("rejects_d.csv")

# Corrected: Directly use the values from the 'timestamp' column
interrej_corrected = df['timestamp'].tolist()

# Compute inter-rejection times
interrej_diffs = [interrej_corrected[i+1] - interrej_corrected[i] for i in range(len(interrej_corrected)-1)]

# Create bins with a width of 0.05
index = max(interrej_diffs)
bins = [i*0.05 for i in range(int(index//0.05 + 1))]
values = [0] * len(bins)

# Populate the values based on inter-rejection times
for i in range(len(interrej_diffs)):
    values[int(interrej_diffs[i]//0.05)] += 1/len(interrej_diffs)

# Plot the values against the bins
plt.figure(figsize=(10, 6))
plt.plot(bins, values, drawstyle='steps-post')
plt.xlabel("Inter-Rejection Time")
plt.ylabel("Frequency")
plt.title("Distribution of Inter-Rejection Times (Using 'timestamp' Column)")
plt.grid(True)
plt.tight_layout()
plt.show()

len(interrej_diffs)
```
